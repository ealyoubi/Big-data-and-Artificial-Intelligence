```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flags_data= pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/flags/flag.data',sep=",", names = ['name','landmass','zone','area','population','language','religion','bars','stripes','colours','red','green','blue','gold','white','black','orange','mainhue','circles','crosses','saltires','quarters','sunstars','crescent','triangle','icon','animate','text','topleft','botright']) 
flags_data
```

```python
flags_data_new= flags_data.loc[0:20,['name','area','population','language','religion','bars','stripes','colours']]
flags_data_new 
```

```python
#Line Chart
```

```python
#sns.lineplot(data = flags_data_new.bars)
sns.lineplot(data=flags_data_new, x=flags_data_new.bars, y=flags_data_new.stripes)
plt.xlabel("bars");
plt.title("Bars and Stripes in Flags")
```

Explanation: bars: Number of vertical bars in the flag
stripes: Number of horizontal stripes in the flag

```python
# Bar Chart
```

```python
# Set the width and height of the figure
sns.barplot(y=flags_data_new['name'], x=flags_data_new['area'])
plt.title("Area of Countries")
```

Explanation: we can see from the barchart that Australila is the largest area comparing with other contries presented in the bar chart. 

```python
#Scatter Plot
```

```python
sns.scatterplot(x=flags_data['religion'], y=flags_data['landmass'])
plt.title("Religion and Landmass")
```

religion: 0=Catholic, 1=Other Christian, 2=Muslim, 3=Buddhist, 4=Hindu, 5=Ethnic, 6=Marxist, 7=Others
landmass: 1=N.America, 2=S.America, 3=Europe, 4=Africa, 4=Asia, 6=Oceania

Explanation: the plot shows that there are diversity in religions in Africa

```python
#Pie Chart
nameofcountries= ['Afghanistan','Albania','Algeria','American-Samoa','Andorra','Angola','Anguilla']
Population = [16,3,20,0,0,7,0]
plt.figure(figsize=(20,10));
plt.pie(Population , labels= nameofcountries)
plt.title("Countries and Population")
plt.show()
 
```

Explanation: the Pie chart shows that Algeria has the biggest population comparing with presented contries

```python
#Histogram
plt.hist(flags_data_new['colours'], bins = 5 , align='right', color='purple', edgecolor='black')
plt.xlabel('colours')
plt.ylabel('index of the Country')
plt.title('Colours in Flags')
 
```

Explanation: the histogram shows that most countries' flags have three colors.

```python
#Boxplot
```

```python
boxplot = flags_data.boxplot(column=['colours'])
```

Explanation:we can see from the boxplot that the miumum number of clors in the flag is 5 and the maximum is 7 with out the outlier values.
